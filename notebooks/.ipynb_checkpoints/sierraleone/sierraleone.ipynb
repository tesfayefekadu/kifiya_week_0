```python
def find_outliers_iqr(column):
    Q1 = column.quantile(0.25)  # 25th percentile
    Q3 = column.quantile(0.75)  # 75th percentile
    IQR = Q3 - Q1               # Interquartile Range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return column[(column < lower_bound) | (column > upper_bound)]
outliers_moda = find_outliers_iqr(data['ModA'])
outliers_modb = find_outliers_iqr(data['ModB'])
outliers_ws = find_outliers_iqr(data['WS'])
outliers_WSgust = find_outliers_iqr(data['WSgust'])
print("Outliers in ModA:")
print(outliers_moda)
print("Outliers in ModB:")
print(outliers_modb)
print("Outliers in WS:")
print(outliers_ws)
print("Outliers in WSgust:")
print(outliers_WSgust)
```

```python
# Select relevant columns
solar_radiation = ['GHI', 'DNI', 'DHI']
temperature_measures = ['TModA', 'TModB']
wind_conditions = ['WS', 'WSgust', 'WD']

# Combine all relevant columns for different analyses
solar_temp_columns = solar_radiation + temperature_measures
wind_solar_columns = wind_conditions + solar_radiation
```

```python
# Heatmap of Correlation Matrix
# Compute correlation matrix for solar radiation and temperature measures
# Correlation Analysis Between Solar Radiation and Temperature Measures
corr_matrix = data[solar_temp_columns].corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap
sns.heatmap(corr_matrix, annot=True, cmap=cmap, fmt=".2f", linewidths=.5)

plt.title('Correlation Heatmap: Solar Radiation and Temperature Measures')
plt.show()
```

```python
# Pair Plot
# A pair plot visualizes pairwise relationships in the dataset, including scatter plots and histograms.
# Create a pair plot for solar radiation and temperature measures
sns.pairplot(data[solar_temp_columns].dropna(), diag_kind='hist', corner=True)
plt.suptitle('Pair Plot: Solar Radiation and Temperature Measures', y=1.02)
plt.show()
```

```python
# Correlation Analysis Between Wind Conditions and Solar Irradiance
from pandas.plotting import scatter_matrix

# Select data for scatter matrix
scatter_data = data[wind_solar_columns].dropna()

# Create scatter matrix
scatter_matrix(scatter_data, figsize=(12, 12), diagonal='kde', alpha=0.5)

plt.suptitle('Scatter Matrix: Wind Conditions and Solar Irradiance', y=1.02)
plt.show()
```

```python
# List of columns to analyze
columns_to_analyze = ['GHI', 'DNI', 'DHI', 'WS', 'TModA', 'TModB']

# Calculate Z-scores for each column and identify outliers
outliers = {}

for column in columns_to_analyze:
    # Calculate Z-scores
    data[f'{column}_zscore'] = stats.zscore(data[column].dropna())  # dropna() to handle missing values
    
    # Identify outliers (Z-scores greater than 3 or less than -3)
    outliers[column] = data[(data[f'{column}_zscore'].abs() > 3)]

    # Display Z-scores and outliers
    print(f"Outliers for {column}:\n", outliers[column][[column, f'{column}_zscore']])

# Optionally, save the data with Z-scores and outliers to a new CSV file
data.to_csv('data_with_zscores.csv', index=False)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 10))

# Plot histograms of Z-scores for each column
for i, column in enumerate(columns_to_analyze, 1):
    plt.subplot(2, 3, i)
    plt.hist(data[f'{column}_zscore'].dropna(), bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Z-Scores of {column}')
    plt.xlabel('Z-Score')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
